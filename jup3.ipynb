```python
import pip

def import_or_install(package):
    try:
        __import__(package)
    except ImportError:
        pip.main(['install', package])

import_or_install("numpy")
import_or_install("pandas")

import numpy as np
import pandas as pd
from numpy.random import random

np.random.seed(5)


df = pd.DataFrame(np.random.rand(5, 5), columns=['V', 'W', 'X', 'Y', 'Z'], index=['A', 'B', 'C', 'D', 'E'])
print(df)

df["new"] = df["W"] + df["Y"]
df.drop("new", axis=1, inplace=True)
df = df.drop("E", axis=0)

print("greater than 0")
print(df[df>0]) #parts of df greater than 0

print("w & y")
print(df[df["W"] > 0]["Y"]) #parts of W in df greater than 0 and Y

print("y > 1")
print(df[(df["W"]>0) & (df["Y"] > 1)]) #parts of W in df greater than 0 and Y greater than 1
```

```python
import pip

def import_or_install(package):
    try:
        __import__(package)
    except ImportError:
        pip.main(['install', package])

import_or_install("numpy")
import_or_install("pandas")

import numpy as np
import pandas as pd
from numpy.random import random

#example of hierarchial multi-level index 

outside = ["G1", "G1", "G1", "G2", "G2", "G2"]
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)

print(hier_index)
df = pd.DataFrame (np.random.randn(6,2),index=hier_index, columns=["A","B"])
df

print("loc")
print(df.loc["G1"].loc[1])

df.index.names = ["Group", "Num"]
df
```
