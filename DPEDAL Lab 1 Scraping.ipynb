```python
import pip

def import_or_install(package):
    try:
        __import__(package)
    except ImportError:
        pip.main(['install', package])

import_or_install("numpy")
import_or_install("pandas")
import_or_install("bs4")
import_or_install("requests")

import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests as re

url = "https://www.sitpune.edu.in/"

response = re.get(url)

print(response)
```

```python
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
soup.find('title')
```

```python
soup.find('title').string
```

```python
h1_tags = soup.find_all('h1')
print(h1_tags)
```

```python
h2_tags = soup.find_all('h2')
print(h2_tags)
```

```python
h3_tags = soup.find_all('h2')
for h3 in h3_tags:
    print(h3.text)
```

```python
a_tags = soup.find_all('a')
len(a_tags)

list_of_links = []
for a in a_tags:
    link = a.get('href')
    if link != None:
        list_of_links.append(link)

len(list_of_links)

df_links = pd.DataFrame(data=list_of_links, columns=['Links'])

df_links.to_csv(r'sit_links.csv')
```

```python
li_tags = soup.find_all('li')
for li in li_tags:
    print(li.text)
```

```python
link = soup.select('img[src^="https://www.sitpune.edu.in/assets/images"]')
img_links = []

directory = 'scraped_images/'
```

```python
for img in link:
    img_links.append(img['src'])
```

```python
for index, img_link in enumerate(img_links):
    img_data = re.get(img_link).content
    with open(f'{directory}/'+str(index+1)+'.jpg', 'wb+') as f:
        f.write(img_data)
```
