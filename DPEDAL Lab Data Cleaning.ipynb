## Data Imputation

## fill pre_movie score with mean value of pre_movie_score

```python
df['pre_movie_score']
```

```python
df['pre_movie_score'].mean()
```

```python
df['pre_movie_score'] = df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```

```python
df
```

## replaced null values through median

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df
```

```python
df['post_movie_score'] = df['post_movie_score'].fillna(df['post_movie_score'].median())
```
